# 1.4.7 Canny Edge Detection

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('tree1.jpg',0)
blur = cv2.GaussianBlur(img, (5,5),3)
edges = cv2.Canny(img,100,200)
edges_gaussianblur = cv2.Canny(blur, 100, 200)

plt.rcParams["figure.figsize"] = (20,10)

plt.subplot(221),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(222),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.subplot(223),plt.imshow(blur,cmap = 'gray')
plt.title('Gaussian blur'), plt.xticks([]), plt.yticks([])

plt.subplot(224),plt.imshow(edges_gaussianblur,cmap = 'gray')
plt.title('Gaussian blur - edge'), plt.xticks([]), plt.yticks([])

plt.show()
```
