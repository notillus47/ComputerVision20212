# Control Flow

```python
# if statements(as expected)
x=2
if x ==3:
    print("X equals 3.")
elif x == 2:
    print("X equals 2.")
else:
    print("X equals something else.")
print("This is outside the 'if'")
```

```python
x = 3
while x < 5:
    print(x, "still in the loop")
    x += 1
```

```python
x = 6
while x < 5:
    print(x, "still in the loop")
```

```python
#Break and continue
```

```python
x = 1
while 1:
    print(x)
    x += 1
    if(x==20):
        break
        
```

```python
x = 1
while 1:
    print(x)
    x += 1
    continue
    if(x==20):
        break
        
```
