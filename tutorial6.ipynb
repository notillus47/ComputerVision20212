# For Loops

```python
# for <item> in <collection>:
#    <statements>
```

```python
item = [1, 2, 3, 4]
for i in item:
    print(i)
```

```python
# For loops and the range() function
# range(5) returns [0,1,2,3,4]
for x in range(5):
    print (x)
```

```python
#Abuse of the range() function
for (i,item) in enumerate(item):
    print (i, item)
```

```python
# Avoid
for i in range(len(item)):
    print(item[i])
```
