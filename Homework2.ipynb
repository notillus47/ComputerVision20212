```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture ('tracking1.avi')
```

```python
red = 150
l = np.array ([red, 50, 50])
u = np.array ([red+40, 255, 255])
```

```python
while (cap.isOpened()) :
    
    ret, frame = cap.read()
    
    hsv = cv2.cvtColor (frame, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange (hsv, l, u)
    res = cv2.bitwise_and (frame, frame, mask = mask)

    cv2.imshow ('frame', frame)
    cv2.imshow ('tracking', res)
    
    k = cv2.waitKey (42) & 0xFF
    if (k == 27) :
        break
        
    if not ret :
        break

cap.release()      
cv2.destroyAllWindows ()
```
