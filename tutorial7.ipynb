# Generating Lists using "List Comprehensions"

```python
# [expression for name in list]
```

```python
li = [3, 6, 2, 7]
[elem*2 for elem in li]
```

```python
li = [('a', 1), ('b', 2), ('c', 7)]
[n*3 for (x,n)in li]
```

```python
#Filter determines whether expression is performed on each member of the list
# When processing each element of list, first check if it satisfies the filter condition.
# If the filter condition returns False, that element is omitted from the list before the list comprehensionn is evaluated
```

```python
li = [3, 6, 2, 7, 1, 9]
[elem*2 for elem in li if elem>4]
```

```python
# Nested List comprehensions
```

```python
li = [3, 2, 4, 1]
[elem*2 for elem in[item+1 for item in li]]
```
